#### 3. í…ì„œ ë°ì´í„°ì™€ ë°°ì¹˜ ë§Œë“¤ê¸°

```python
# ëª¨ë¸ í•™ìŠµì— í•„ìš”í•œ ë°ì´í„°ë¥¼ í…ì„œ ë°ì´í„°ë¡œ ë³€í™˜í•˜ëŠ” í´ë˜ìŠ¤ (data_loader íŒŒì¼ ì°¸ê³ )
# ì…ë ¥ê°’(x)ê³¼ ëª©í‘œê°’(y)ì˜ ë‹¨ìœ„ í˜•íƒœë¥¼ ë§ì¶°ì•¼ ì¶”í›„ ê³¼ì •ì— ìš©ì´í•˜ë‹¤.(ì˜ˆì¸¡ ê²°ê³¼ì™€ ì •ë‹µì˜ ë¹„êµ ì‹œ, loss ê³„ì‚° ì‹œ ë“±ë“±)
class TensorData(Dataset):
  def __init__(self, x_data, y_data):
    self.x_data = torch.FloatTensor(x_data) # float íƒ€ì…ì˜ í…ì„œë¡œ ë°”ê¿”ì¤€ë‹¤.
    self.y_data = torch.FloatTensor(y_data)
    self.len = self.y_data.shape[0] # [í–‰, ì—´] ì¤‘ í–‰ì— í•´ë‹¹í•œë‹¤.

  # (ì…ë ¥, íƒ€ê²Ÿ) íŠœí”Œ í˜•íƒœë¡œ ë°ì´í„°ë¥¼ ë°˜í™˜
  # PyTorchëŠ” ì´ ë©”ì„œë“œë¥¼ í†µí•´ DataLoaderì—ì„œ í•œ ìƒ˜í”Œì”© ê°€ì ¸ì˜¤ê²Œ ëœë‹¤.
  def __getitem__(self, index):
    return self.x_data[index], self.y_data[index]
  
  # ë°ì´í„°ì…‹ì˜ ì „ì²´ ê¸¸ì´(ìƒ˜í”Œ ìˆ˜)ë¥¼ ë¦¬í„´í•œë‹¤.
  def __len__(self):
    return self.len
```

```python
# train_test_split : ë°ì´í„°ë¥¼ ëœë¤í•˜ê²Œ ì„ê³ (shuffle), ì „ì²´ ë°ì´í„°ë¥¼ í•™ìŠµ ë°ì´í„°ì™€ í‰ê°€ ë°ì´í„°ë¡œ ë‚˜ëˆˆë‹¤.
# test size = 0.5 -> ì „ì²´ ë°ì´í„° ì¤‘ 50%ë¥¼ testìš© ë°ì´í„°ë¡œ, ë‚˜ë¨¸ì§€ 50%ëŠ” í•™ìŠµìš© ë°ì´í„°ë¡œ ë‚˜ëˆ„ê² ë‹¤ëŠ” ì˜ë¯¸ë‹¤.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5)

# TensorData: PyTorchì—ì„œ ì‚¬ìš©í•  ìˆ˜ ìˆëŠ” Dataset í˜•íƒœë¡œ ë³€í™˜
# DataLoader(ë¬¶ìŒ): Datasetì„ batch ë‹¨ìœ„ë¡œ êº¼ë‚¼ ìˆ˜ ìˆê²Œ ë§Œë“¤ì–´ì¤Œ
# í•™ìŠµ ë°ì´í„° ë°°ì¹˜ í˜•íƒœë¡œ êµ¬ì¶•í•˜ê¸°
train_sets = TensorData(X_train, Y_train)
train_loader = DataLoader(train_sets, batch_size=32, shuffle=True)

# ì‹œí—˜ ë°ì´í„° ë°°ì¹˜ í˜•íƒœë¡œ êµ¬ì¶•í•˜ê¸°
test_sets = TensorData(X_test, Y_test)
test_loader = DataLoader(test_sets, batch_size=32, shuffle=False)

# ë§Œì•½ batch_sizeê°€ 2 ë¼ë©´ ì•„ë˜ì²˜ëŸ¼ ë‘ ìƒ˜í”Œì´ í•œ batchì— ë¬¶ì—¬ì„œ ë“¤ì–´ì˜¤ê²Œ ëœë‹¤.
# ì…ë ¥ x_batch: tensor([[ 3., 30.],
#                      [ 1., 10.]])
# ì •ë‹µ y_batch: tensor([[300.],
#                      [100.]])
```

#### 4. ëª¨ë¸ êµ¬ì¶•

Regressor ëª¨ë¸(íšŒê·€ ëª¨ë¸)ëŠ” ì…ë ¥ì¸µ(ë…¸ë“œ 13ê°œ), 2ê°œì˜ ì€ë‹‰ì¸µ(50ê°œ, 30ê°œ), ì¶œë ¥ì¸µ(1ê°œ)ìœ¼ë¡œ êµ¬ì„±í•œë‹¤.
* Regression = íšŒê·€. ì¦‰, ìˆ«ìë¥¼ ì˜ˆì¸¡í•˜ëŠ” ëª¨ë¸ì„ ëœ»í•œë‹¤. ìˆ«ì ë²”ìœ„ ì•ˆì— ì •í™•íˆ ì–¼ë§ˆë‚˜ ê°€ê¹Œìš´ì§€ ë§ì¶”ëŠ” ê²Œ í•µì‹¬ì´ë‹¤. ex. ì§‘ê°’, ì£¼ì‹  

ë°ì´í„°ì˜ ë³€ìˆ˜ëŠ” 13ê°œì´ë¯€ë¡œ ì…ë ¥ì¸µì˜ ë…¸ë“œëŠ” 13ê°œê°€ ë˜ê³  ì¶œë ¥ì¸µì€ ì§‘ ê°’ì¸ ë‹¨ì¼ ê°’ì„ ì¶”ì¶œí•˜ëŠ” ê²ƒì´ë¯€ë¡œ 1ê°œê°€ ëœë‹¤. 

ì€ë‹‰ì¸µì— ëŒ€í•´ì„œëŠ” ì‹¤í—˜ì„ í•˜ë©´ì„œ íŠœë‹í•  ìˆ˜ ìˆë‹¤.

```python
# nn.Moduleì„ ìƒì†ë°›ì•„ì„œ Pytorch ëª¨ë¸ êµ¬ì¡°ë¥¼ ë”°ë¦„
class Regressor(nn.Module):
  def __init__(self):
    super().__init__() # nn.Moduleì˜ ì´ˆê¸°í™” í•¨ìˆ˜ë¥¼ í˜¸ì¶œ -> PyTorch ë‚´ë¶€ì—ì„œ ëª¨ë¸ íŒŒë¼ë¯¸í„°ë¥¼ ì¶”ì í•˜ê³  ì—…ë°ì´íŠ¸í•˜ë ¤ë©´ ê¼­ í•„ìš”í•˜ë‹¤.
    self.fc1 = nn.Linear(13, 50, bias=True) # ì…ë ¥ì¸µ(13) -> ì€ë‹‰ì¸µ1(50)ìœ¼ë¡œ ê°€ëŠ” ì—°ì‚°
    self.fc2 = nn.Linear(50, 30, bias=True) # ì€ë‹‰ì¸µ1(50) -> ì€ë‹‰ì¸µ2(30)ìœ¼ë¡œ ê°€ëŠ” ì—°ì‚°
    self.fc3 = nn.Linear(30, 1, bias=True) # ì€ë‹‰ì¸µ2(30) -> ì¶œë ¥ì¸µ(1)ìœ¼ë¡œ ê°€ëŠ” ì—°ì‚°
    self.dropout = nn.Dropout(0.2) # ì—°ì‚°ì´ ë  ë•Œë§ˆë‹¤ 20%ì˜ ë¹„ìœ¨ë¡œ ëœë¤í•˜ê²Œ ë…¸ë“œë¥¼ ì—†ì•¤ë‹¤.

  # ì‹¤ì œ ì—°ì‚° ìˆœì„œ ì •ì˜
  # ëª¨ë¸ì— ì…ë ¥ xê°€ ë“¤ì–´ì™”ì„ ë•Œ ê³„ì‚°ì´ ì–´ë–»ê²Œ í˜ëŸ¬ê°ˆì§€ë¥¼ ì •ì˜
  # â—ï¸ ì°¸ê³ ë¡œ Regressor(x)ì™€ ê°™ì´ ëª¨ë¸ ì¸ìŠ¤í„´ìŠ¤ë¥¼ í•¨ìˆ˜ì²˜ëŸ¼ ì“°ë©´, forward()ëŠ” ìë™ìœ¼ë¡œ í˜¸ì¶œë˜ë©° ì‹¤í–‰ëœë‹¤.
  def forward(self, x):
    x = F.relu(self.fc1(x)) # self.fc1(x) : ì…ë ¥ì„ ì²« ë²ˆì§¸ Linear ê³„ì¸µ(13 -> 50)ì— í†µê³¼ì‹œí‚¤ê³ , F.relu() -> í™œì„±í™” í•¨ìˆ˜ ReLUë¥¼ ì ìš©í•œë‹¤. (ìŒìˆ˜ëŠ” 0, ì–‘ìˆ˜ëŠ” ê·¸ëŒ€ë¡œ)   
    x = self.dropout(F.relu(self.fc2(x))) # ë‘ ë²ˆì§¸ Linear ê³„ì¸µ(50 -> 30)ì— í†µê³¼ì‹œí‚¤ê³ , ì€ë‹‰ì¸µ2ì—ì„œ ë“œëì•„ì›ƒì„ ì ìš©í•œë‹¤. (ì¦‰, 30ê°œì˜ 20%ì¸ 6ê°œì˜ ë…¸ë“œê°€ ê³„ì‚°ì—ì„œ ì œì™¸ëœë‹¤.)
    x = F.relu(self.fc3(x)) # Linear ê³„ì‚° í›„ í™œì„±í™” í•¨ìˆ˜ ReLUë¥¼ ì ìš©í•œë‹¤. (ê°€ê²© ì˜ˆì¸¡) 
        
    return x
    
# ì£¼ì˜ ì‚¬í•­
# ë“œëì•„ì›ƒì€ ê³¼ì í•©(overfitting)ì„ ë°©ì§€í•˜ê¸° ìœ„í•´ ë…¸ë“œì˜ ì¼ë¶€ë¥¼ ë°°ì œí•˜ê³  ê³„ì‚°í•˜ëŠ” ë°©ì‹ì´ê¸° ë•Œë¬¸ì— ì ˆëŒ€ë¡œ ì¶œë ¥ì¸µì— ì‚¬ìš©í•´ì„œëŠ” ì•ˆ ëœë‹¤.
```

##### íšŒê·€(Regression) vs ë¶„ë¥˜(Classification)

| í•­ëª©             | íšŒê·€ (Regression)                     | ë¶„ë¥˜ (Classification)                  |
|------------------|----------------------------------------|-----------------------------------------|
| ì˜ˆì¸¡ ê°’           | ì—°ì†ì ì¸ ìˆ«ì                          | ê³ ì •ëœ í´ë˜ìŠ¤(label) ì¤‘ í•˜ë‚˜           |
| ì˜ˆì‹œ             | ì§‘ê°’ ì˜ˆì¸¡: `250000`, `301000`         | ì´ë©”ì¼ ë¶„ë¥˜: `ìŠ¤íŒ¸`, `ì¼ë°˜`             |
| ì¶œë ¥ í˜•íƒœ         | í•˜ë‚˜ì˜ ì‹¤ìˆ˜ ê°’ (ì˜ˆ: `nn.Linear(..., 1)`) | í´ë˜ìŠ¤ ìˆ˜ë§Œí¼ ì¶œë ¥ (ì˜ˆ: `nn.Linear(..., 3)`) |
| ëŒ€í‘œ ëª¨ë¸ ì´ë¦„     | `Regressor`                            | `Classifier`                            |
| ì†ì‹¤ í•¨ìˆ˜ (Loss) | `nn.MSELoss()`, `nn.L1Loss()`          | `nn.CrossEntropyLoss()` ë“±              |
| í‰ê°€ ì§€í‘œ         | MSE, MAE, RMSE, RÂ² ë“±                  | ì •í™•ë„(Accuracy), ì •ë°€ë„(Precision), ì¬í˜„ìœ¨ ë“± |
| ì‹œê°ì  ìš”ì•½       | ğŸ  ì…ë ¥: ë©´ì , ë°© ìˆ˜ â†’ ì¶œë ¥: `356000`     | ğŸ“§ ì…ë ¥: ì´ë©”ì¼ ë‚´ìš© â†’ ì¶œë ¥: `ìŠ¤íŒ¸ or ì¼ë°˜`  |

#### 5. ëª¨ë¸, ì†ì‹¤í•¨ìˆ˜, ìµœì í™” ë°©ë²• ì„ ì–¸

```python
model = Regressor()

# ì†ì‹¤í•¨ìˆ˜
# MSE(í‰ê·  ì œê³± ì˜¤ì°¨) = mean((ì˜ˆì¸¡ê°’ - ì‹¤ì œê°’)^2)
criterion = nn.MSELoss()

# ëª¨ë¸ì„ í•™ìŠµì‹œí‚¬ ë•Œ íŒŒë¼ë¯¸í„°ë¥¼ ì–´ë–»ê²Œ ì—…ë°ì´íŠ¸í• ì§€ ê²°ì •í•˜ëŠ” ë¶€ë¶„
# lr: í•™ìŠµë¥ ì´ë‹¤. í•œ ë²ˆì˜ í•™ìŠµ(step)ì—ì„œ ì–¼ë§ˆë‚˜ í¬ê²Œ ê°€ì¤‘ì¹˜ë¥¼ ì¡°ì •í• ì§€ë¥¼ ì˜ë¯¸í•œë‹¤. í¬ë©´ ë¹¨ë¦¬ í•™ìŠµë˜ì§€ë§Œ ë¶ˆì•ˆì •í•  ìˆ˜ ìˆê³ , ì‘ìœ¼ë©´ ì•ˆì •ì ì´ì§€ë§Œ ëŠë¦´ ìˆ˜ ìˆë‹¤.
# weight_decay
# - L2 ì •ê·œí™”ì˜ ì •ë„, penalty ê°’ì„ ì˜ë¯¸í•œë‹¤. 1e-2:0.01, 1e-3:0.001,.., 1e-6
# - ê°’ì´ í´ìˆ˜ë¡ í° ê°€ì¤‘ì¹˜ì— ë” ë§ì€ íŒ¨ë„í‹°ë¥¼ ì¤€ë‹¤. ê°€ì¤‘ì¹˜ = í•™ìŠµë˜ëŠ” íŒŒë¼ë¯¸í„° 
# - ëª¨ë¸ì´ ë„ˆë¬´ ë³µì¡í•˜ê²Œ í•™ìŠµí•˜ëŠ” ê²ƒ(overfitting)ì„ ë§‰ê¸° ìœ„í•´ ì‚¬ìš© ëœë‹¤.

# Adam: ê¸°ì¡´ì˜ ê°’ì„ ëˆ„ì í•´ ê°€ëŠ” ì•Œê³ ë¦¬ì¦˜
optimizer = optim.Adam(
  model.parameters(), 
  lr=1e-2, # = 0.01
  weight_decay=1e-7
)
```

```python
# â—ï¸
# í”„ë¦¬íŠ¸ë ˆì¸ ëœ ëª¨ë¸ì´ ì—†ë‹¤. = (pret = 0)
# ì¤‘ê°„ì— ëŠê²¨ì„œ ë‹¤ì‹œ ì´ì–´ì„œ ëŒë¦¬ëŠ” ê²½ìš°(./models/reg4-2.pt ì´ ìƒê¸´ ì´í›„)ë¼ë©´ 1ë¡œ ë°”ê¿”ì¤€ë‹¤. ì €ì¥ëœ íŒŒì¼ì„ ë¶ˆëŸ¬ì™€ì„œ ê°ê° ì •ë³´ë¥¼ ì¬ ì„ ì–¸í•˜ê²Œ ëœë‹¤.
pret = 1

if pret == 1:
    checkpoint = torch.load ('./models/reg4-2.pt')
    model.load_state_dict(checkpoint['model'])
    optimizer.load_state_dict(checkpoint['optimizer'])
    loss_ = checkpoint['loss']
    ep = checkpoint['epoch']
    ls = loss_[-1]

    print(f"epoch={ep}, loss={ls}") # ê²°ê³¼ epoch=215, loss=0.0038022379158064723 ë¡œ 215ë¶€í„° ë‹¤ì‹œ ì‹œì‘í•˜ëŠ” ê²ƒì„ ë³¼ ìˆ˜ ìˆë‹¤.
    # ë‹¤ìŒ epoch ë¶€í„° ì‹œì‘í•´ì•¼ í•˜ë‹ˆê¹Œ 1ì„ ë”í•œ ì—í­ ë¶€í„° ì‹œì‘í•œë‹¤.
    ep = ep + 1

else:
    ep = 0
    ls = 1 # lossê°€ 1 ë¯¸ë§Œì¸ ê²½ìš° ë² ìŠ¤íŠ¸ ë¡œìŠ¤ ê¸°ì¤€ìœ¼ë¡œ ëª¨ë¸ì„ ì €ì¥
    loss_ = [] # ê·¸ë˜í”„ë¥¼ ê·¸ë¦¬ê¸° ìœ„í•œ loss ì €ì¥ìš© ë¦¬ìŠ¤íŠ¸ 
```

#### 6. í•™ìŠµ ì§„í–‰

```python
ls = 1 # ì„ì˜ì˜ ì†ì‹¤ ì„ê³„ê°’
loss_ = [] # ê·¸ë˜í”„ë¥¼ ê·¸ë¦¬ê¸° ìœ„í•œ loss ì €ì¥ìš© ë¦¬ìŠ¤íŠ¸ 
n = len(train_loader)

for epoch in range(2000): # 200ë²ˆ í•™ìŠµì„ ì§„í–‰í•œë‹¤.

  running_loss = 0.0
  
  for data in train_loader: # ë¬´ì‘ìœ„ë¡œ ì„ì¸ 32ê°œ ë°ì´í„°ê°€ ìˆëŠ” ë°°ì¹˜ê°€ í•˜ë‚˜ ì”© ë“¤ì–´ì˜¨ë‹¤.

    inputs, values = data # dataì—ëŠ” X, Yê°€ ë“¤ì–´ìˆë‹¤.

    optimizer.zero_grad() # ìµœì í™” ì´ˆê¸°í™”
    
    outputs = model(inputs) # ëª¨ë¸ì— ì…ë ¥ê°’ ëŒ€ì… í›„ ì˜ˆì¸¡ê°’ ì‚°ì¶œ
    loss = criterion(outputs, values) # ì†ì‹¤ í•¨ìˆ˜ ê³„ì‚°
    loss.backward() # ì†ì‹¤ í•¨ìˆ˜ ê¸°ì¤€ìœ¼ë¡œ ì—­ì „íŒŒ ì„¤ì •, ì†ì‹¤ í•¨ìˆ˜ ê¸°ì¤€ìœ¼ë¡œ ê° ê°€ì¤‘ì¹˜ê°€ ì†ì‹¤ì— ì–¼ë§ˆë‚˜ ì˜í–¥ì„ ì¤¬ëŠ”ì§€ ê³„ì‚°í•˜ëŠ” ë‹¨ê³„ = ì†ì‹¤ ê°’ì˜ gradient(ê¸°ìš¸ê¸°)ë¥¼ ê³„ì‚°í•´ ì €ì¥
    optimizer.step() # ì—­ì „íŒŒë¥¼ ì§„í–‰í•˜ê³  ê°€ì¤‘ì¹˜(íŒŒë¼ë¯¸í„°) ì—…ë°ì´íŠ¸
    
    running_loss += loss.item() # epoch ë§ˆë‹¤ í‰ê·  lossë¥¼ ê³„ì‚°í•˜ê¸° ìœ„í•´ ë°°ì¹˜ lossë¥¼ ë”í•œë‹¤.

  l = running_loss/n
  loss_.append(l) # ì†ì‹¤ í•¨ìˆ˜ë¥¼ ë¦¬ìŠ¤íŠ¸ì— ì¶•ì 

  # ê°€ì¥ ì†ì‹¤ì´ ì‘ì•˜ë˜ ì‹œì ì˜ ì •ë³´ë¥¼ ì €ì¥ (ìµœê³  ì„±ëŠ¥ì´ ì–¸ì œ ì˜€ëŠ”ì§€ ê¸°ë¡)
  if l < ls: # ì¤‘ë‹¨ ì—†ì´ ë°˜ë³µí•˜ë¯€ë¡œ early stopping ê°œë…ì€ ì•„ë‹˜
    ls = l
    ep = epoch
    
    # â­ï¸ torch.save ì•ˆì— ë”•ì…”ë„ˆë¦¬ë¥¼ ë„£ì–´ì„œ ì—¬ëŸ¬ ì •ë³´ë¥¼ ì €ì¥í•  ìˆ˜ ìˆë‹¤.
    torch.save({'epoch': ep,
                'loss': loss_, # ì €ì¥ ëì„ ë•Œ ê¹Œì§€ì˜ loss ë¦¬ìŠ¤íŠ¸
                'model': model.state_dict(),
                'optimizer': optimizer.state_dict()
                }, './models/reg4-2.pt')

        
print('Finished Training')
```

#### 7. ëª¨ë¸ í‰ê°€

```python
# RMSE (Root Mean Squared Error) 
# ì˜ˆì¸¡ ì„±ëŠ¥ì„ í‰ê°€í•  ë•Œ ì‚¬ìš©
def rmse(dataloader):
  with torch.no_grad(): # requires_grad: ë¹„í™œì„±í™” => ì—­ì „íŒŒë¥¼ ë¹„í™œì„±í™”í•˜ì—¬ í‰ê°€í•  ë•ŒëŠ” ëª¨ë¸ì„ ì—…ë°ì´íŠ¸í•˜ì§€ ì•Šì•„ ë¶ˆí•„ìš”í•œ ê³„ì‚°ì„ ë§‰ëŠ”ë‹¤. 
      square_sum = 0
      num_instances = 0
      model.eval() # ëª¨ë¸ì„ í‰ê°€ ëª¨ë“œë¡œ ë°”ê¿ˆ (ì•„ë˜ ì¶”ê°€ ì„¤ëª…)

      for data in dataloader:
          inputs, targets = data
          outputs = model(inputs)
          square_sum += torch.sum((outputs - targets)**2).item()
          num_instances += len(targets)

  model.train()

  return np.sqrt(square_sum/num_instances)  

# í‰ê°€ ì‹œ .eval()ì„ ì‚¬ìš©í•´ì•¼ í•˜ëŠ” ì´ìœ 
# í‰ê°€ ì‹œì—ëŠ” ì •ê·œí™” ê¸°ìˆ ì„ ë°°ì œí•˜ì—¬ ì˜¨ì „í•œ ëª¨ë¸ë¡œ í‰ê°€ë¥¼ í•´ì•¼í•œë‹¤. ë”°ë¼ì„œ .eval()ì„ ì‚¬ìš©í•œë‹¤.
# ì¦‰, ë“œëì•„ì›ƒì´ë‚˜ ë°°ì¹˜ ì •ê·œí™” ë“±ê³¼ ê°™ì´ í•™ìŠµ ì‹œì—ë§Œ ì‚¬ìš©í•˜ëŠ” ê¸°ìˆ ë“¤ì´ ì ìš© ëœ ëª¨ë¸ì€ í‰ê°€ ì‹œì—ëŠ” ë¹„í™œì„±í™” í•´ì•¼í•˜ë©° í•™ìŠµ ì‹œ .train()ì„ ì‚¬ìš©í•œë‹¤.
```

```python
train_rmse = rmse(train_loader) # í•™ìŠµ ë°ì´í„°ì˜ RMSE
test_rmse = rmse(test_loader) # ì‹œí—˜ ë°ì´í„°ì˜ RMSE

print("Train RMSE: %.5f" %train_rmse)
print("Test RMSE: %.5f" %test_rmse)

# ì˜ˆì‹œë¥¼ ìœ„í•œ ë‹¨ìˆœ ë¹„êµì´ë‹¤. ì‹¤ì œ ì—°êµ¬ì—ì„œëŠ” ë””í…Œì¼í•œ ë¹„êµê°€ í•„ìš”í•˜ë‹¤.
# ì˜ˆë¥¼ ë“¤ì–´ 20ë²ˆì˜ í‰ê°€ ê²°ê³¼ì˜ í‰ê· ê³¼ í‘œì¤€í¸ì°¨ë¡œ ìš°ìœ„ ë¹„êµë¥¼ í•˜ê±°ë‚˜ ì‹ ë¢°êµ¬ê°„ì„ ì´ìš©í•œë‹¤.
# ë°ì´í„°ë¥¼ ë¬´ì‘ìœ„ë¡œ ë‚˜ëˆ„ê³  ëª¨ë¸ì˜ ì´ˆê¸°ê°’ë„ random initial parameterë¥¼ ì‚¬ìš©í–ˆê¸° ë•Œë¬¸ì— í•™ìŠµì„ í•  ë•Œ ë§ˆë‹¤ ê²°ê³¼ê°€ ë‹¤ë¥´ê²Œ ë‚˜ì˜¬ ìˆ˜ ìˆë‹¤.
```
