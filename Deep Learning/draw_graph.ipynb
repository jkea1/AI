### 그래프 그리기
데이터 특징, 분석의 결과, 현재 상태 등의 다양한 분석을 위해 시각화는 매우 중요한 요소 중 하나이다. 따라서 여러가지 시각화 라이브러리가 있지만 기본적으로 널리 사용되는 Matplotlib을 알아본다. Matplotlib은 매우 많은 기능을 제공하고 있기 때문에 특별한 그래프나 디테일한 세팅은 구글링을 통해 많이 찾아봐야 한다.

추가적으로 데이터 분석에서는 seaborn 라이브러리도 많이 사용된다.

```python
import numpy as np
from matplotlib import pyplot as plt
```

### 1. Plot

```python
epoch = np.arange(1, 100)
train_loss = 1.5/(epoch)
val_loss = .3+1/(epoch)
```

* 후에 epoch(학습시킨 수), train_loss, val_loss, test_loss 등의 값을 이용한 작업들이 많을텐데, 이를 잘 시각화하는 방법을 알아야 minor tuning, 전략 변경, 문제 파악 등이 수월해 진다.

```python
# 선 그래프 2개 그리기
plt.figure(figsize=(10,5))
plt.plot(train_loss, 'r-*') # red line with stars
plt.plot(val_loss, 'b-o') # blue line with circles
plt.legend(['train', 'validation']) # 오른쪽 위에 범례 추가
plt.title('MNIST') # graph title
plt.xlabel('dk') # x-axis label
plt.ylabel('loss') # y-axis label

plt.show() # 그래프 출력
```

```python
# 한 선에서 다양한 함수 그리기
plt.figure(figsize=(5, 10))
t = np.arange(0., 5., 0.2)
plt.plot(t, t,'ro', t, t**2, 'bs', t, 5*np.sin(t), 'g^') # x, y, graph type
plt.legend(['t (linear)', 't**2 (quadratic)', '5sin(t) (sine)'])

plt.show() # 그래프 출력
```

### 2. Subplot
* 여러가지 그래프를 한 창에 그리고 싶은 경우 사용

```python
names = ['First', 'Business', 'Economy']
prices = [20, 5, 1]

plt.figure(figsize=(15, 5)) # 전체 figure 크기 설정

plt.subplot(131) # 1행 3열의 1번째 subplot
plt.bar(names, prices)
plt.title(f"{names[0]}", fontsize=14)

plt.subplot(132) # 1행 3열의 2번째 subplot
plt.scatter(names, prices) # 산점도
plt.xlabel('class', fontsize=12)

plt.subplot(133) # 1행 3열의 3번째 subplot
plt.plot(names, prices, linewidth=5.0)
plt.suptitle('Ticket Class', fontsize=14)
plt.axis('off')

plt.show()
```
