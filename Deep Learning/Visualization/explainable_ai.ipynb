### CAM(Class Activity Map)
CAM(Class Activity Map)은 설명 가능한 AI 기술 중 하나다. 이미지 내에서 어느 부분이 모델의 의사결정에 큰 영향을 미쳤는지를 시각화하는 기술이다.

Reference: "Learning Deep Features for Discriminative Localization" by B.Zhou et al.,2015

#### 1. 라이브러리 호출

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2

import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import torch.nn as nn
import torch.optim as optim
from tqdm import trange
```

#### 3.데이터 및 모델 생성

```python
# GPU vs CPU
# 현재 가능한 장치를 확인한다.
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")

print(f'{device} is available')
```

```python
# 10 classes: airplane, bird, car, cat, deer, dog, horse, monkey, ship, truck

# Compose -> 전처리 과정가 차례대로 될 수 있도록 함
transform = transforms.Compose([transforms.Resize(128), # 원래 96X96 인 STL10을 CAM 모델에 맞추기 위해 리사이즈
                                transforms.ToTensor(), # 텐선 (C, H, W) 형태로 변환하고 동시에 픽셀 값도 [0, 1]범위로 정규화
                                transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))]) # R, G, B 채널 각각에 대해 0.5, 표준 편차 0.5로 정규화, 픽셀 값이 [0, 1] -> [-1, 1] 범위로 바뀜
trainset = torchvision.datasets.STL10(root='./data', split='train', download=True, transform=transform) # 96x96
trainloader = torch.utils.data.DataLoader(trainset, batch_size=32, shuffle=True)
```

```python
model = torchvision.models.resnet18(weights='DEFAULT') # ResNet-18 구조를 쓰되, ImageNet으로 이미 똑똑해진 모델을 불러온다.

model.conv1 = nn.Conv2d(3, 64, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False)
model.fc = nn.Linear(512, 10)
model = model.to(device)
```

#### 3. 모델 학습

```python
criterion = nn.CrossEntropyLoss() # 다중 클래스 분류용 손실함 수 
optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=1e-2)
```

```python
num_epochs = 10
ls = 2
pbar = trange(num_epochs)

for epoch in pbar:
    correct = 0
    total = 0
    running_loss = 0.0

    for data in trainloader:
        inputs, labels = data[0].to(device), data[1].to(device)
          
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels) # 얼마나 틀렸는지 숫자로 계산
        loss.backward() # 어디서 틀렸는지 확인
        optimizer.step() # 다음에 더 잘 맞히도록 가중치 업데이트(틀린 쪽은 줄이고, 잘 한 쪽은 유지하거나 조금 늘린다.)

        running_loss += loss.item()
        _, predicted = torch.max(outputs.detach(), 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    cost = running_loss / len(trainloader)   
    acc = 100*correct / total

    if cost < ls:
        ls = cost
        torch.save(model.state_dict(), './models/stl10_resnet18.pth')    

    pbar.set_postfix({'loss' : cost, 'train acc' : acc}) 
```

```python
correct = 0
total = 0

with torch.no_grad():
    model.eval()

    for data in trainloader:
        images, labels = data[0].to(device), data[1].to(device)
        outputs = model(images) # 학습시키 모델을 사용한다. -> 각 클래스에 대해서 얼마나 확신하는지 결과를 낸다.
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print('Accuracy of the network on the train images: %d %%' % (100 * correct / total))
```

#### 5. CAM 모델 구축

```python
# Visualize feature maps
activation = {} # 특정 레이어에서 나오는 feature map을 저장하기 위한 전역 변수

def get_activation(name):
    def hook(model, input, output):
        activation[name] = output.detach() # 해당 레이어의 output 값을 저장, detach 하여 
        
    return hook
```

```python
def cam(model, trainset, img_sample, img_size):
    model.eval()

    with torch.no_grad(): # requires_grad 비활성화
        model.layer4[1].bn2.register_forward_hook(get_activation('final')) # feature extraction의 마지막 feature map 구하기
        data, label = trainset[img_sample] # 이미지 한 장과 라벨 불러오기
        data.unsqueeze_(0) # 4차원 3차원 [피쳐수, 너비, 높이] -> [1, 피쳐수, 너비, 높이]
        output = model(data.to(device)) # forward pass
        _, prediction = torch.max(output, 1)
        act = activation['final'].squeeze() # 4차원 [1,피쳐수 ,너비, 높이] -> 3차원 [피쳐수 ,너비, 높이]
        w = model.fc.weight # classifer의 가중치 불러오기
        
        # 각 채널의 feature map을 해당 class에 대한 중요도(w)로 weighted sum 한다.
        for idx in range(act.size(0)): # CAM 연산
            if idx == 0:
                tmp = act[idx] * w[prediction.item()][idx] 
            else:
                tmp += act[idx] * w[prediction.item()][idx]

        # 모든 이미지 팍셀값을 0~255로 스케일하기 
        normalized_cam = tmp.cpu().numpy()
        normalized_cam = (normalized_cam - np.min(normalized_cam)) / (np.max(normalized_cam) - np.min(normalized_cam)) # CAM의 값을 0~1로 정규화
        original_img = np.uint8((data[0][0] / 2 + 0.5) * 255) # 0~255 스테일로 변환(기존 정규된 이미지를 되돌린다.)

        # 원본 이미지 사이즈로 리사이즈
        cam_img = cv2.resize(np.uint8(normalized_cam * 255), dsize=(img_size, img_size)) # CAM 이미지를 원본 이미지와 같은 크기로 리사이즈하고 0~255로 스케일함ㅁ
        
    return cam_img, original_img
```

```python
def plot_cam(model, trainset, img_size, start):
    end = start + 20
    fig, axs = plt.subplots(2, (end - start + 1) // 2, figsize=(20, 5))
    fig.subplots_adjust(hspace=.01, wspace=.01)
    axs = axs.ravel()

    # CAM과 원본 이미지를 각각 불러와서, 원본 위에 CAM을 반투명으로 겹쳐서 시각화 한다.
    for i in range(start, end):
        cam_img, original_img = cam(model, trainset, i, img_size)

        axs[i - start].imshow(original_img, cmap='gray')
        axs[i - start].imshow(cam_img, cmap='jet', alpha=.5)
        axs[i - start].axis('off')

    plt.show()
    fig.savefig('./results/cam.png')
```

```python
plot_cam(model, trainset, 128, 10) # 128x128 크기로, 10~29번째 이미지
```
